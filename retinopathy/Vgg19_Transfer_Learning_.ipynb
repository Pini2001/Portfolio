## CNN

A convolutional neural network (CNN) is a deep learning algorithm model that uses a neural network architecture for image processing. This neural network architecture includes three main stages: preprocessing, feature extraction and classification.

The CNN is composed of several layers of neurons: convolution layers, pooling layers, and fully connected layers, which work together to perform these three stages of image processing.

It is important to note that, although the basic process of a convolutional neural network involves applying convolutional operations and assigning labels to an input image, the entire process is much more complex than it appears. Therefore, the steps involved in this process will be detailed in detail below.

First, the input image is transformed into a matrix of numerical values (pixels) representing the color intensity in each of the image channels (red, green and blue) with dimensions (width x height x color channels).

Second, the matrix of numerical values is fed into the convolution layer, where relevant image features, such as edges, patterns and shapes, are extracted and irrelevant features are removed. This is done through the use of filters (kernels), filters are small size matrices such as 3x3 or 5x5, which are used to detect specific patterns in the input image during the convolution phase.

Each filter scrolls through the matrix of image values, scanning a portion of the image matrix, and an element multiplication operation is performed between the values in the filter matrix and the corresponding values in the image section covered by the filter.

Then, all the results of the multiplication are summed and the resulting value is placed in a new matrix called feature map (or activation maps). This step is repeated until the entire image is scanned.

As more convolution layers are added to the network, the filters become more complex and specialize on more specific features, allowing the network to learn to detect increasingly abstract and complex features in the image. Each filter has a set of associated weights that are used to multiply the pixel section of the corresponding image. The weights determine which image features are important for the classification task. These weights are initialized with random values and are adjusted during network training, they are very important because they determine which image features are relevant to the classification task.

Once the filters are applied, a feature map containing information about the important properties of the image is obtained. This feature map is used as input for the next convolutional layer, the grouping layers. Grouping layers are used in CNNs to reduce the dimension of the feature map, using techniques such as subsampling or clustering of feature values, both techniques aim to reduce the feature matrix.


To recapitulate, both techniques reduce the amount of information in the feature map, which in turn reduces the amount of network parameters and makes the information more manageable for subsequent layers. This stage is very important to prevent overfitting and reduce the complexity of the model.

The last layer of the convolutional neural network is the fully connected layer. In this layer, all the features learned from the previous convolutional layers are taken and combined to form a one-dimensional vector. This vector is then fed to an output layer that performs the classification task. The output layer can use a softmax (multi-class output variable) or sigmoid (binary output variable) activation function to compute the classification probabilities of the images fed to the CNN. These probabilities are assigned to specific classes so that the image is classified into one of them.

## Challenges

The implementation of a CNN (or convnet) poses a number of challenges to be taken into account.

One of them is the need for large amounts of high quality data, since DL models require a large collection of data to be able to learn and detect patterns in the input data, so data is essential for their correct operation. The analogy that can be used to describe the relationship between data and DL is that data is like the fuel that powers the machine learning engine, in the same way that gasoline is the fuel that powers a car. Without sufficient and quality data, the DL model cannot function properly, just as a car cannot move without fuel.

Another major challenge is the need for computational resources to train and run DL models. Neural networks are complex and can have many layers, which means they need millions of parameters to learn during training. This can require a large amount of memory and a fast processor to design and train these models.

Finally, creating a CNN from scratch takes a lot of effort in terms of time and energy. The complexity of the model and the long computation times required, which can last from hours to days, require a great deal of dedication on the part of the programmer to properly design and train the network.

## Transfer Learning VGG19

To address the above challenges, techniques such as the use of transfer models, commonly known as pre-trained networks, have been developed. These models take advantage of knowledge that is generated from a large data set and apply it to a related learning task.

In pre-trained models, the convolutional layers have already been trained on large datasets, and useful image features have been extracted and optimized (the feature maps and weights). Therefore, there is no need to retrain those layers and they can be directly used in a specific classification task. Only the clustering layers and the final fully connected layers need to be trained to adapt the model to the specific classification task.

This can save a lot of time and resources compared to training a model from scratch, as such large datasets are not needed. With a smaller dataset, the final fully connected layers and grouping layers can be trained to adapt the model to the specific task, obtaining good accuracy in image classification.

In the following, we will explain the steps necessary to create a VGG-19 based transfer model that allows us to classify images for a diagnosis:

           -1. Load the pre-trained VGG19 model with the ImageNet weights and without the final dense layer.
           -2. Freeze the convolutional layers of the base model to not allow the weights to be updated during training.
           -3. Add a new dense layer on top of the base model for the classification task.
           -4. Create the final model by specifying the inputs and outputs.
           -5. Compile the model by specifying the loss function, the optimizer and the metric.
           -6. Split the data to evaluate and train the model, as well as define the transformations to be applied to the images during training and model validation.
           -7. Train the model with our data using the preprocessed data and the added output layers.
           -8. Evaluate the model with a test data set to measure its accuracy and performance.


The VGG19 model is composed of a total of 22 layers, of which 19 are convolutional layers and the last 3 are dense (fully connected) layers at the end of the model. When creating a transfer model with VGG19, only the 19 convolutional layers containing the pre-trained weights from the ImageNet database are loaded.

Of those 19 layers, 16 are convolution layers and 3 are pooling layers. The dense layers are not loaded as they are designed for the classification task for which the model was originally trained (from ImageNet). After the convolutional layers are loaded, they are frozen to avoid modifying the weights of these layers and to maintain the features learned from the ImageNet images. Then, an average clustering layer is added to reduce the dimensionality of the feature maps. Next, a hidden layer of a single dense layer (Dense) with 100 relu activation neurons is added. This helps the network to detect more complex and specific, nonlinear patterns of features in the reduced maps. Finally, a Dense output layer with a single neuron with the problem-specific activation function is added.

Once we have defined the grouping and fully connected (output) layers we want to add to the pre-trained model, we create the final model with these layers embedded.

Next, the loss function (which measures the discrepancy between the true labels), the optimizer (algorithm used to adjust the neural network weights during training) and the metrics (which evaluate the performance of the model) are configured.

## Preprocess

Image preprocessing in Deep Learning uses techniques such as data augmentation to increase the variability of the training dataset, normalization to avoid saturation of pixel values and center pixel values at zero to improve model efficiency, resizing to standardize image size, and batch creation to speed up training. These techniques are implemented using different methods, such as horizontal flipping, DL model-specific preprocessing functions, image resizing, and batching.

Detailed description of the most common techniques and methods used during the preprocessing process in deep learning models:

* **Horizontal flipping**: This is a data augmentation technique used in DL that generates variations of the original images by rotating them on a vertical axis to create a new image that is a mirror image of the original from another angle. The goal of this technique is to increase the diversity of the training data set, creating images similar to the original but with some differences. By using a wider variety of data, it helps the model learn more efficiently by preventing overfitting and improving its ability to generalize to new data.

* **Specific function in keras**: This function applies a series of transformations so that the images are in a suitable format for the neural network, applying the following transformations.
               -1. It subtracts the average color of the image in RGB format, where each pixel is composed of three color components: red (R), green (G) and blue (B), and each component can take an integer value between 0 and 255. The RGB average of an image is calculated by adding the R, G and B values of all pixels in the image and dividing the result by the total number of pixels.
               -2. The color channels are rearranged so that they are in the proper order for the neural network. For example, the VGG-19 neural network expects the color channels in BGR order instead of RGB.
               -3. In order to reduce the scale of the input data and improve the stability and efficiency of the neural network training process, the pixel values are normalized to be in the range of [-1, 1] which helps to avoid saturation or nonlinear activation problems in the neural network layers.

  Both the subtraction of the mean of the RGB values and the normalization of the pixel values can change the appearance of the image by affecting the color balance, making the image appear more blue or yellow depending on the mean values of the original colors.

* **Image resizing**: Used to reduce the amount of data in a high-resolution image, speed up its processing and homogenize the data size. DL models require a specific input size to function properly by having a fixed number of parameters and layers that fit a given input size. The convolution layer of a neural network (CNN) expects a specific input image size, such as 224x224 pixels. The new image size is achieved by modifying the pixel resolution (number of pixels in the image width and height).

* **Batching**: This technique, also called batching, consists of dividing the training dataset into smaller batches to process them separately and not the entire dataset at once. This has a number of advantages such as reduced memory usage on the GPU, improvements in computational efficiency and the ability to process large data sets that would not fit in the computer's memory. It also allows for more precise tuning of the neural network weights by updating the weights after each batch processed.

For DR diagnosis using deep learning models, the preprocessing techniques proposed in the theory are used. First, the horizontal flipping technique is applied to augment the training data. Then, a specific preprocessing for the VGG19 neural network is used, which consists in the subtraction of the RGB mean of the images, the rearrangement of the color channels to be in the correct order for the neural network and the normalization of the pixel values of the images. This preprocessing is performed to prepare the images for feeding into the VGG19 neural network. 59Once the images have been transformed into a format suitable for the network, they are resized to a specific input size, in this case 224x224 pixels. Finally, the batch creation technique is used to divide the training dataset into smaller batches to reduce memory usage and improve computational efficiency during model training.

```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.applications.vgg19 import preprocess_input

def generators(shape, preprocessing, train_csv):
    train_datagen = ImageDataGenerator(
        preprocessing_function=preprocessing,
        horizontal_flip=True,
        validation_split=0.25
    )

    # Load the CSV file with image filenames and their corresponding labels
    df = pd.read_csv(train_csv)
    df['id_code'] = df['id_code'].apply(lambda x: x + '.png')
    df['label'] = np.where(df['diagnosis'] == 0, 'No', 'Si')

    # Create the data generators
    train_generator = train_datagen.flow_from_dataframe(
        dataframe=df,
        directory='/content/DR',
        x_col='id_code',
        y_col='label',
        target_size=(224, 224),
        batch_size=30,
        class_mode='binary',
        subset='training'
    )

    val_generator = train_datagen.flow_from_dataframe(
        dataframe=df,
        directory='/content/DR',
        x_col='id_code',
        y_col='label',
        target_size=(224, 224),
        batch_size=30,
        class_mode='binary',
        subset='validation'
    )

    return train_generator, val_generator
```

## VGG19

Load the pre-trained VGG19 model with ImageNet weights, excluding the top (fully connected) dense layers.

```python
base_model = VGG19(weights='imagenet', include_top=False, input_shape = (224,224, 3))
print('VGG19 Descargado')
resumen(base_model)
```

Freeze the convolutional layers of the base model.

```python
for layer in base_model.layers:
    layer.trainable = False
```

Add a new dense layer on top of the base model for our binary classification task.

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
#GlobalAveragePooling2D() tends to work better for tasks where the spatial location of the features is not as important
#such as image classificationas it allows the model to focus more on the global patterns in the image.
# Flatten() can be more suitable for tasks where the spatial location of the features is important, such as object detection, as it preserves the spatial information in the output.
# one hidden layers
x = Dense(100, activation='relu')(x) # small dataset it is generally advisable to start with a smaller number of neurons in the intermediate layer to avoid overfitting
#x = Dense(128, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)  # for binary classification between patients with diabetic retinopathy or not.
```

Create the final model by specifying the inputs and outputs.

```python
# creating the full model:
model = Model(inputs=base_model.input, outputs=predictions)
resumen(model)
```

Compile the model specifying the loss function, optimizer, and metric.

```python
import tensorflow.keras.backend as K

def sensitivity(y_true, y_pred):
    """Calculate sensitivity, also known as recall or true positive rate."""
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    return true_positives / (possible_positives + K.epsilon())

def specificity(y_true, y_pred):
    """Calculate specificity, also known as true negative rate."""
    true_negatives = K.sum(K.round(K.clip((1-y_true) * (1-y_pred), 0, 1)))
    possible_negatives = K.sum(K.round(K.clip(1-y_true, 0, 1)))
    return true_negatives / (possible_negatives + K.epsilon())

def precision(y_true, y_pred):
    """Calculate precision, also known as positive predictive value."""
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    return true_positives / (predicted_positives + K.epsilon())

```

```python
model.compile(loss='binary_crossentropy',
              optimizer=Adam(lr=0.001),
              metrics=['accuracy', sensitivity, specificity, precision])
```

```python
def plot_history(history, yrange):
    '''Plot loss and accuracy as a function of the epoch,
    for the training and validation datasets.
    '''
    acc = history.history['accuracy'] #['acc']
    #acc = history.history['acc']
    val_acc = history.history['val_accuracy']
    #val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    # Get number of epochs
    epochs = range(len(acc))

    # Plot training and validation accuracy per epoch
    plt.plot(epochs, acc)
    plt.plot(epochs, val_acc)
    plt.title('Training and validation accuracy')
    plt.ylim(yrange)

    # Plot training and validation loss per epoch
    plt.figure()

    plt.plot(epochs, loss)
    plt.plot(epochs, val_loss)
    plt.title('Training and validation loss')

    plt.show()
```

Train the model with our data.

```python
history = model.fit_generator(train_generator, epochs=5, validation_data=val_generator, workers=10)
plot_history(history, yrange=(0.9,1))
```

```python
import matplotlib.pyplot as plt

# Extract the values of the validation metrics from the history object
val_specificity = history.history['val_specificity']
val_sensitivity = history.history['val_sensitivity']
val_accuracy = history.history['val_accuracy']
val_precision = history.history['val_precision']

# Create a new figure with subplots for each metric
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))

# Plot the values for each metric
axes[0].plot(range(1, 6), val_specificity, 'bo-')
axes[0].set_title('Validation Specificity')
axes[0].yaxis.set_major_formatter('{:.2f}'.format)
axes[1].plot(range(1, 6), val_sensitivity, 'go-')
axes[1].set_title('Validation Sensitivity')
axes[1].yaxis.set_major_formatter('{:.2f}'.format)
axes[2].plot(range(1, 6), val_accuracy, 'ro-')
axes[2].set_title('Validation Accuracy')
axes[2].yaxis.set_major_formatter('{:.2f}'.format)
axes[3].plot(range(1, 6), val_precision, 'ro-')
axes[3].set_title('Validation Precision')
axes[3].yaxis.set_major_formatter('{:.2f}'.format)

# Set common axis labels and a main title
fig.suptitle('Validation Metrics over 5 Epochs')
fig.text(0.5, 0.04, 'Epoch', ha='center')
fig.text(0.04, 0.5, 'Score', va='center', rotation='vertical')

# Show the plot
plt.show()
```
